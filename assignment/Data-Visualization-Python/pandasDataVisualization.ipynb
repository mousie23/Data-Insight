# Importing Datasets

```python
# First, we'll import pandas, a data processing and CSV file 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="white", color_codes=True)
```

```python
# Next, we'll load the Iris flower dataset
iris = pd.read_csv("Iris.csv")
#to check the first five rows of our datas 
iris.head()
```

```python
# To check each species numbers
iris["Species"].value_counts()
```

# Scatter Plot

```python
iris.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
```

```python
# We can also make a boxplot with Pandas on each feature split out by species
iris.drop("Id", axis=1).boxplot(by="Species", figsize=(12, 6))
```

# Scatter plot using Seaborn

```python
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris, size=5)
```

```python
# Here we cannot identify which one belongs to which species so
# We'll use seaborn's FacetGrid to color the scatterplot by species and also add legend
sns.FacetGrid(iris, hue="Species", size=5) \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()
```

# Boxplot using seaborn

```python
# We can look at an individual feature in Seaborn through a boxplot
sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
```

```python
# One way we can extend this plot is adding a layer of individual points on top of it through Seaborn's striplot
# 
# We'll use jitter=True so that all the points don't fall in single vertical lines above the species
#
# Saving the resulting axes as ax each time causes the resulting plot to be shown
# on top of the previous axes
ax = sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
ax = sns.stripplot(x="Species", y="PetalLengthCm", data=iris, jitter=True, edgecolor="gray")
```
